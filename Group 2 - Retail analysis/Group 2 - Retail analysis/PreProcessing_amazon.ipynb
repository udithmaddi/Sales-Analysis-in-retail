```python
data['Agent_Age'] = data['Agent_Age'].fillna(data['Agent_Age'].mean())
```

```python
data['Agent_Rating'] = pd.to_numeric(data['Agent_Rating'], errors='coerce')
```

```python
data['Agent_Rating'] = data['Agent_Rating'].fillna(data['Agent_Rating'].mean())
```

```python
data['Order_Date'] = pd.to_datetime(data['Order_Date'], errors='coerce')
```

```python
data['Order_Time'] = pd.to_datetime(data['Order_Time'], format='%H:%M:%S', errors='coerce')
data['Pickup_Time'] = pd.to_datetime(data['Pickup_Time'], format='%H:%M:%S', errors='coerce')
```

```python
data['Weather'] = data['Weather'].str.lower().str.strip()
```

```python
data.drop('Order_ID', axis=1, inplace=True)
```

```python
data.drop('Store_Latitude',axis=1, inplace=True)
```

```python
data.drop('Store_Longitude',axis=1, inplace=True)
```

```python
data.to_csv('PP_amazon_data.csv',index=False)
```
